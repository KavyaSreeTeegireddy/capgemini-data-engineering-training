```python
from pyspark.sql import SparkSession
```

```python
from pyspark.sql.functions import *
spark = SparkSession.builder.appName("Spark DataFrames").getOrCreate()
df = spark.read.options(header='True', inferSchema='True').csv('/Volumes/workspace/default/cap_vol/C1.csv')
df.show()
```

```python
df_clean = df.withColumn(
    "Sales",
    when(col("Sales") == "blank", 0)
    .otherwise(col("Sales"))
)
```

```python
df_clean = df_clean.withColumn(
    "Category",
    when(col("Category") == "Blank","Unknown")
    .otherwise(col("Category"))
)
display(df_clean)
```

```python
df_clean= df_clean.withColumn(
    "Country",
    when(col("Country")== "india", "India")
    .when(col("Country") == "New York", "USA")
    .otherwise(col("Country"))
)
df_clean.show()
```

```python
df_clean = df_clean.filter(col("CustomerID") != 104)
df_clean.show()
```

```python
df_clean = df_clean.withColumn(
    "JoinDate",
    when(col("JoinDate") == "blank", "01-01-2023")
    .otherwise(col("JoinDate"))
)
```

```python
df_clean = df_clean.withColumn(
    "JoinDate",
    when(
        col("CustomerID") == 106,
        date_format(
            to_date(col("JoinDate"), "MM-dd-yyyy"),
            "dd-MM-yyyy"
        )
    ).otherwise(col("JoinDate"))
)

df_clean.show()
```
