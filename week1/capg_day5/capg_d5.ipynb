```python

from pyspark.sql.functions import *
df.select(col("emp_id"),col("name"),col("address")).display()
```

```python

df_1 = df.withColumn(
    "salary_category",
    when(col("salary") >= 7000, "High")
    .when(col("salary") >= 6000, "Medium")
    .otherwise("Low")
)
df_1.show()
```

```python
data1 = [
    (101, "Alice", 5000, "Street 1", "USA", "alice@gmail.com"),
    (102, "Bob", 6000, "Street 2", "India", "bob@gmail.com"),
    (103, "Charlie", 5500, "Street 3", "UK", "charlie@gmail.com"),
    (104, "David", 7000, "Street 4", "USA", "david@gmail.com")
]

data2 = [(105, "Eve", 6500, "Street 5", "UK", "eve@gmail.com"),
    (106, "Frank", 4800, "Street 6", "India", "frank@gmail.com"),
    (107, "Grace", 7200, "Street 7", "USA", "grace@gmail.com"),]
cols = ["emp_id", "name", "salary", "address", "loc", "email"]

df1 = spark.createDataFrame(data1, cols)
df2 = spark.createDataFrame(data2, cols)
df_union = df1.union(df2)
df_union.show()
```
